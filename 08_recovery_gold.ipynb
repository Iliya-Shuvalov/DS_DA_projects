# Восстановление золота из руды

Цель работы: подготовить прототип модели машинного обучения. Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. Для построения модели используются данные с параметрами добычи и очистки. Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.
  
Описание данных:  
Rougher feed — исходное сырье  
Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant  
Xanthate — ксантогенат (промотер, или активатор флотации)  
Sulphate — сульфат (на данном производстве сульфид натрия)  
Depressant — депрессант (силикат натрия)  
Rougher process (англ. «грубый процесс») — флотация  
Rougher tails — отвальные хвосты  
Float banks — флотационная установка  
Cleaner process — очистка  
Rougher Au — черновой концентрат золота  
Final Au — финальный концентрат золота  
  
Параметры этапов:  
air amount — объём воздуха  
fluid levels — уровень жидкости  
feed size — размер гранул сырья  
feed rate — скорость подачи  
rougher — флотация  
primary_cleaner — первичная очистка  
secondary_cleaner — вторичная очистка  
final — финальные характеристики  
input — параметры сырья  
output — параметры продукта  
state — параметры, характеризующие текущее состояние этапа  
calculation — расчётные характеристики

### Проверка эффективности обогащения

Вычислим эффективность обогащения на обучающей выборке для признака rougher.output.recovery. Найдём MAE между расчётным значением и значением признака:

**recovery = ((C(F-T))/(FC-T))*100%**

- С - доля золота концентрате после фильтрации/очитски (rougher.output.concentrate_au)
- F - доля золота в сырье/концентрате до флотации/очистки (rougher.input.feed_au)
- T - доля золота в отвальных хвостах после фильтрации/очистки (rougher.output.tail_au)

```python
c = df_train['rougher.output.concentrate_au'] #доля золота концентрате после фильтрации/очитски
f = df_train['rougher.input.feed_au'] #доля золота в сырье/концентрате до флотации/очистки
t = df_train['rougher.output.tail_au'] #доля золота в отвальных хвостах после фильтрации/очистки

recovery = ((c*(f-t))/(f*(c-t))) * 100

print('MAE между расчётным значением и значением признака =', mean_absolute_error(df_train['rougher.output.recovery'], recovery))
```

Значение МАЕ близкое к нулю подтверждает правильность расчета эффективности обогащения

```python
#Поскольку соседние по времени параметры похожи, то пропуски в данных заполним интерполяцией 
df_train = df_train.interpolate()
```

```python
df_train.info()
```

Очистим от выбросов данные по концентрациям металлов и солей на различных стадиях очистки:

```python
#Создадим словари {Наименование металла : обозначение в данных} и {наименование этапа очистки : обозначение в данных}
dict_metals = {'Золото':'au', 'Серебро':'ag', 'Свинец':'pb'}
dict_stages = {'Флотация':'rougher.output.', 'Первичная очистка':'primary_cleaner.output.','Вторичная очистка':'final.output.'}
```

```python
INTERVAL_SIGMA = 3 #Оставим данные в пределах 3 сигма (99.73 % данных в случае нормального распределения)
```

```python
for elements in ['sol', 'au', 'ag', 'pb']:
    for stage in dict_stages:
        up_lim =  df_train[f'{dict_stages[stage]}concentrate_{elements}'].mean() + \
        df_train[f'{dict_stages[stage]}concentrate_{elements}'].std()*INTERVAL_SIGMA #Верхний предел для среза данных
        
        down_lim = df_train[f'{dict_stages[stage]}concentrate_{elements}'].mean() - \
        df_train[f'{dict_stages[stage]}concentrate_{elements}'].std()*INTERVAL_SIGMA #Нижний предел для среза данных
        
        df_train = df_train[(df_train[f'{dict_stages[stage]}concentrate_{elements}'] > down_lim) & \
                            (df_train[f'{dict_stages[stage]}concentrate_{elements}'] < up_lim)]
```

Аналогично очистим от выбросов данные по размеров гранул сырья (оставим данные в пределах 3 сигма):

```python
up_lim =  df_train['primary_cleaner.input.feed_size'].mean() + df_train['primary_cleaner.input.feed_size'].std()*INTERVAL_SIGMA
down_lim = df_train['primary_cleaner.input.feed_size'].mean() - df_train['primary_cleaner.input.feed_size'].std()*INTERVAL_SIGMA

df_train = df_train[(df_train['primary_cleaner.input.feed_size'] > down_lim) &\
                            (df_train['primary_cleaner.input.feed_size'] < up_lim)]
```

```python
up_lim =  df_train['rougher.input.feed_size'].mean() + df_train['rougher.input.feed_size'].std()*INTERVAL_SIGMA
down_lim = df_train['rougher.input.feed_size'].mean() - df_train['rougher.input.feed_size'].std()*INTERVAL_SIGMA

df_train = df_train[(df_train['rougher.input.feed_size'] > down_lim) &\
                            (df_train['rougher.input.feed_size'] < up_lim)]
```

### Функция для вычисления sMAPE

```python
#Функция для расчета sMAPE. Принимает на вход target и predicted модели, возвращает значение sMAPE  
def smape(target,predicted):
    return np.mean(200*abs(target - predicted)/(abs(target)+abs(predicted)))
```

```python
#Функция для расчета итогового sMAPE  
def smape_result(rougher, final):
    return(0.25*rougher + 0.75*final)
```
